```python
import locale
import os

import pandas as pd
```

```python
# Download the 16 csv files from https://drive.google.com/drive/folders/1d0idkOmMIXabj7ajYhvkitMMHnH_woSN
df = pd.concat(
    [
        pd.read_csv("../../data/raw/hackathon_13_09_2024/" + f)
        for f in os.listdir("../../data/raw/hackathon_13_09_2024/")
    ]
)
```

```python
# Check the content of the data
with pd.option_context("display.max_columns", None):
    display(df)
```

```python
# Rename the columns and keep the most relevant ones
columns_renaming = {
    "ID": "id",
    "Start": "start",
    "Plaintext": "text",
    "Channel Name": "channel_name",
    "Channel Title": "channel_title",
    "Channel Radio": "channel_is_radio",
    "Channel Program Type": "channel_program_type",
    "Channel Program": "channel_program",
    "Theme": "themes",
    "Keywords With Timestamp": "keywords",
    "Number Of Keywords": "num_keywords",
}
df = df.rename(columns=columns_renaming)
df = df[list(columns_renaming.values())].set_index("id")
df
```

```python
# Parse the start date
locale.setlocale(locale.LC_TIME, "fr_FR.UTF-8")

df["start"] = pd.to_datetime(df["start"], format="%d %B, %Y, %H:%M")
df
```

```python
# There are a few duplicates, remove them
print(f"There are {df.index.duplicated().sum()} duplicates")
df = df[~df.index.duplicated(keep="last")]
```

```python
# Merge the channels
display(df[["channel_name", "channel_title"]].drop_duplicates().reset_index(drop=True))

# I think we can just keep the "channel_name" column
# Maybe we would like to distinguish "i>Télé" and "CNews"?
df = df.copy()
df["channel_name"] = df["channel_name"].replace({"franceinfotv": "france_info"})

# Drop redundant channel title
df = df.drop(columns="channel_title")

print(f"There are {df["channel_name"].nunique()} channels")
```

```python
# We have a few NaN values, let's remove them
display(df.isna().sum(axis=0))
df = df.dropna(how="any")
```

```python
# Check start and end dates for each channel
start_end_dates = df.groupby("channel_name")["start"].agg(["min", "max"])
start_end_dates["min"] = start_end_dates["min"].dt.date
start_end_dates["max"] = start_end_dates["max"].dt.date
start_end_dates.sort_values(["min", "max"])
```

```python
# Final df
df
```
