```python
# Analysis

print("TP")
print("---")
common_models_tp = tp_first_model.intersection(new_model_positives)
print("common:", len(common_models_tp))

print("new   :", len(new_tp_id))

new_model_missed_tp = tp_first_model.difference(new_model_positives)
print("missed:", len(new_model_missed_tp))
print("")

print("FP")
print("---")

new_model_fp = new_model_positives.difference(new_tp_id).difference(common_models_tp)
common_models_fp = fp_first_model.intersection(new_model_fp)

print("new prompt:", len(new_model_fp))
print("old prompt:", len(fp_first_model))

print("common:", len(common_models_fp))

print("")
print("NEW PROMPT")
print("---")
recall_new_model = round((len(common_models_tp) + len(new_tp_id)) / (len(common_models_tp) + len(new_tp_id) + len(new_model_missed_tp)) *100,2)
print("Recall:", recall_new_model)
precision_new_model = round((len(common_models_tp) + len(new_tp_id)) / len(new_model_fp) *100,2)
print("Precision:", precision_new_model)

print("")
print("OLD PROMPT")
print("---")
recall_old_model = round(len(tp_first_model) / (len(tp_first_model) + len(new_tp_id)) *100,2)
print("Recall:", recall_old_model)
precision_old_model = round(len(tp_first_model) / (len(tp_first_model) + len(fp_first_model)) *100,2)
print("Precision:", precision_old_model)
```
