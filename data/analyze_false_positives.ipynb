```python
import pandas as pd
import plotly.express as px
from datasets import load_dataset
```

```python
dataset = load_dataset("DataForGood/climateguard")
df = pd.concat([dataset["train"].to_pandas(), dataset["test"].to_pandas()])
df_fp = df.loc[df.misinformation==False]
df_tp = df.loc[df.misinformation]
display(df_fp.head())
display(df_tp.head())
```

```python
import plotly.graph_objects as go
df_counts_channel = pd.merge(
    df_fp.groupby(["channel"]).agg(counts_fp=("id", "count")),
    df_tp.groupby(["channel"]).agg(counts_tp=("id", "count")),
    left_index=True,
    right_index=True,
    how="outer",
).fillna(0)


px.bar(
    df_counts_channel,
    y=["counts_fp", "counts_tp"],
    barmode="group",
)
```

```python
import asyncio
import ast
from pydantic import BaseModel
from openai import AsyncClient

async_client = AsyncClient()


class TranscriptType(BaseModel):
    number_of_speakers: int
    conversation_type: str

async def get_discussion_type(idx, transcript, semaphore=None):
    if semaphore:
        await semaphore.acquire()
    try:
        chat_response = await async_client.responses.parse(
            model="gpt-5-nano",
            reasoning={"effort": "minimal"},
            #temperature=0,
            input=[{"role": "developer", "content": "You are an assistant helping editors to aggregate tv transcripts into categories of debate and monologue."},
                {
                    "role": "user",
                    "content": f"""
                            Given a transcript from a tv or radio program, it is your duty to determine whether the transcript contains a monologue or a debate.
                            A monologue is when only one speaker can be identified and a debate is when two or more speakers are identified.

                            Transcript: {transcript}
                            Return the answer in json format with the keys 'number_of_speakers' and 'conversation_type', where conversation_type
                            is one of debate and monologue.
                            """
                },
            ],
            text_format=TranscriptType,
        )
    finally:
        if semaphore:
            semaphore.release()
    try:
        return idx, ast.literal_eval(chat_response.output_text)
    except SyntaxError:
        return idx, str(chat_response.output_text)
```

```python
await get_discussion_type(0, df.plaintext[0])
```

```python
from tqdm.asyncio import tqdm

semaphore = asyncio.Semaphore(20)
tasks = [
    get_discussion_type(
        idx,
        row["plaintext"],
        semaphore=semaphore
    ) for idx, row in df_fp.reset_index().iterrows()
]
conv_results_fp = await tqdm.gather(*tasks)
```

```python
conv_results_fp = sorted(conv_results_fp, key=lambda x: x[0])
```

```python
conv_results_fp_df = pd.DataFrame([record[1] for record in conv_results_fp])
df_fp = pd.concat([df_fp.reset_index(), conv_results_fp_df], axis=1)
```

```python
semaphore = asyncio.Semaphore(20)
tasks = [
    get_discussion_type(
        idx,
        row["plaintext"],
        semaphore=semaphore
    ) for idx, row in df_tp.reset_index().iterrows()
]
conv_results_tp = await tqdm.gather(*tasks)
conv_results_tp = sorted(conv_results_tp, key=lambda x: x[0])
conv_results_tp_df = pd.DataFrame([record[1] for record in conv_results_tp])
df_tp = pd.concat([df_tp.reset_index(), conv_results_tp_df], axis=1)
```

```python
df_conv_type = pd.merge(
    df_fp.groupby("conversation_type").agg(counts_fp=("id", "count")),
    df_tp.groupby("conversation_type").agg(counts_tp=("id", "count")),
    left_index=True,
    right_index=True,
)
```

```python
px.bar(df_conv_type, y=["counts_fp", "counts_tp"], barmode="group")
```

```python
df_fp["hour"] = pd.to_datetime(df_fp.start).dt.tz_convert('Europe/Paris').dt.hour
df_tp["hour"] = pd.to_datetime(df_tp.start).dt.tz_convert('Europe/Paris').dt.hour

df_hour = pd.merge(
    df_fp.groupby("hour").agg(counts_fp=("id", "count")),
    df_tp.groupby("hour").agg(counts_tp=("id", "count")),
    left_index=True,
    right_index=True,
)
px.bar(df_hour, y=["counts_fp", "counts_tp"], barmode="group")
```

```python
df_fp["weekday"] = pd.to_datetime(df_fp.start).dt.tz_convert('Europe/Paris').dt.weekday
df_tp["weekday"] = pd.to_datetime(df_tp.start).dt.tz_convert('Europe/Paris').dt.weekday

df_hour = pd.merge(
    df_fp.groupby("weekday").agg(counts_fp=("id", "count")),
    df_tp.groupby("weekday").agg(counts_tp=("id", "count")),
    left_index=True,
    right_index=True,
)
px.bar(df_hour, y=["counts_fp", "counts_tp"], barmode="group")
```
