```python
import ast

def rewrite_transcript(transcript):
    chat_response = openai.beta.chat.completions.parse(
        #temperature=0,
        model="gpt-5-mini",
        messages=[{"role": "system", "content": "You are an assistant helping editors working on tv transcripts."},
            {
                "role": "user",
                "content": f"""
                        You are given a transcript from a tv program, which can be messy and have inexact ponctuation. Rewrite it in proper form.
                        Transcript: {transcript}
                        Return on the rewrite in the following format:
                        <transcript>Insert transcript here..</transcript>
                        """
            },
        ]
    )
    return chat_response.choices[0].message.content
```

```python
import re

def parse_re(text: str, pattern: str, verbose=False):
    """
    Parse the input text using the provided regular expression pattern.

    Parameters:
    - text (str): The input text to be parsed.
    - pattern (str): The regular expression pattern used for searching and matching in the text.
    - verbose (bool, optional): If True, print the entire text if no match is found. Default is False.

    Returns:
    - str: The matched pattern from the text if found; otherwise, an empty string.
    """

    match = re.findall(pattern, text, re.DOTALL)
    if match:
        return match
    if verbose:
        print(text)
    return ""
```

```python
def get_labels(given_labels, transcript):
    chat_response = openai.beta.chat.completions.parse(
        model="gpt-5-mini",
        #temperature=0,
        messages=[{"role": "system", "content": "You are an assistant helping editors to aggregate tv transcripts regarding the climate into themes."},
            {
                "role": "user",
                "content": f"""
                        Given the themes, and under a text classification scenario where editors aim to generalize a tv transcript into an abstract theme expressed in a single sentence, can all these texts be matched to the given themes? 
                        If the transcript does not match any of the themes, please generate a meaningful new theme name.
                        
                        Labels: {given_labels}
                        Transcript: {transcript}
                        You should NOT return meaningless themes names such as ’new_theme_1’ or ’unknown_topic_1’ and only return the new themes, please return the theme in a python list.

                        """
            },
        ]
    )
    return chat_response.choices[0].message.content
```

```python
import asyncio
from openai import AsyncClient
async_client = AsyncClient()


async def get_labels_async(given_labels, transcript, semaphore=None):
    if semaphore:
        await semaphore.acquire()
    try:
        chat_response = await async_client.responses.create(
            model="gpt-5-nano",
            reasoning={"effort": "minimal"},
            #temperature=0,
            input=[{"role": "developer", "content": "You are an assistant helping editors to aggregate tv transcripts regarding the climate into themes."},
                {
                    "role": "user",
                    "content": f"""
                            Given the themes, and under a text classification scenario where editors aim to generalize a tv transcript into an abstract theme expressed in a single sentence, can all these texts be matched to the given themes? 
                            If the transcript does not match any of the themes, please generate a meaningful new theme name.
                            
                            Themes: {given_labels}
                            Transcript: {transcript}
                            You should NOT return meaningless themes names such as ’new_theme_1’ or ’unknown_topic_1’ and only return the new themes, please return the theme in a python list.
                            If the theme of the transcript is described by one of the given themes just return the corresponding theme in a pyhton list. Nothing else.

                            """
                },
            ]
        )
    finally:
        if semaphore:
            semaphore.release()
    try:
        return ast.literal_eval(chat_response.output_text)
    except SyntaxError:
        return str(chat_response.output_text)
```

```python
await get_labels_async(start_labels, df_fp.plaintext_whisper.tolist()[0], None)
```

```python
import math
from tqdm.asyncio import tqdm as atqdm
from tqdm import tqdm

batch_size = 15
n_batches = int(math.ceil(len(df_fp.plaintext_whisper) / batch_size))
labels = start_labels
for i in tqdm(range(n_batches)):
    tqdm.write(f"Processing batch number {i+1}")
    sem = asyncio.Semaphore(10)
    try:
        generated_labels = await atqdm.gather(
            *[
                get_labels_async(
                    labels, 
                    transcript, 
                    sem
                ) for transcript in df_fp.plaintext_whisper.tolist()[i*batch_size:(i+1)*batch_size]
            ] 
        )
        for label in generated_labels:
            if isinstance(label, str):
                labels.append(label)
            elif isinstance(label, list):
                labels.extend(label)
            elif label is None:
                labels.append("")
            else:
                raise TypeError(f"Label must be list or string, instead found {type(label)} for {label}")
        labels = list(set(labels))
    finally:
        del sem
```
